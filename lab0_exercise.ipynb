```python
"""
Exercise 5: Pandas Data Analysis
Analyze a dataset of student performance.
"""

import pandas as pd
import numpy as np

# Create sample dataset
np.random.seed(42)
n_students = 200

data = {
    "student_id": range(1000, 1000 + n_students),
    "major": np.random.choice(["CS", "Math", "Physics", "Biology"], n_students),
    "year": np.random.choice([1, 2, 3, 4], n_students),
    "exam_score": np.random.normal(75, 10, n_students).clip(0, 100),
    "assignments_completed": np.random.randint(0, 11, n_students),
    "hours_studied": np.random.normal(15, 5, n_students).clip(1, 40),
}


df = pd.DataFrame(data)

# Introduce some NaN values
df.loc[np.random.choice(n_students, 10), "exam_score"] = np.nan
df.loc[np.random.choice(n_students, 5), "hours_studied"] = np.nan

# Task 1: Data Cleaning and Exploration (10 points)
# TODO: Display basic information about the dataset
# TODO: Identify and count missing values
# TODO: Fill missing exam_score with the mean score for the student's major
# TODO: Fill missing hours_studied with the median for the student's year

print(df.info())
print(df.describe())
print(df.isnull().sum())
df["exam_score"] = df.groupby("major")["exam_score"].transform(
    lambda x: x.fillna(x.mean())
)
df["hours_studied"] = df.groupby("year")["hours_studied"].transform(
    lambda x: x.fillna(x.median())
)

# Task 2: Analysis (10 points)
# TODO: Calculate and display the average exam_score by major
# TODO: Find the major with the highest average exam_score
# TODO: Calculate the correlation between hours_studied and exam_score
# TODO: Create a new column 'performance' with categories:
#       'Excellent' (>90), 'Good' (80-90), 'Average' (70-80), 'Needs Improvement' (<70)

avg_exam_score = df.groupby("major")["exam_score"].mean()
print(avg_exam_score)
highavg_exam_score = avg_exam_score.max()
correlation = df["hours_studied"].corr(df["exam_score"])


def categorize(score):
    if score > 90:
        return "Excellent"
    elif score >= 80:
        return "Good"
    elif score >= 70:
        return "Average"
    else:
        return "Needs Improvement"


df["performance"] = df["exam_score"].apply(categorize)


# Task 3: Advanced Analysis (10 points)
# TODO: For each major and year combination, calculate:
#       - Number of students
#       - Average exam score
#       - Average hours studied
# TODO: Identify top 5 students based on exam_score (handle ties appropriately)
# TODO: Create a pivot table showing average exam_score by major (rows) and year (columns)

grouped_students = df.groupby(["major", "year"]).agg(
    num_students=("student_id", "count"),
    avg_exam_score=("exam_score", "mean"),
    avg_hours_studied=("hours_studied", "mean"),
)
top_students = df.nlargest(5, "exam_score", keep="all")[
    ["student_id", "major", "year", "exam_score"]
]
pivot = df.pivot_table(
    values="exam_score", index="major", columns="year", aggfunc="mean"
)
print(pivot)
```

```python
"""
Exercise 7: Integration Challenge
Combine NumPy, Pandas, and Matplotlib to solve a mini data science problem.
"""

# Scenario: You're analyzing customer data for an e-commerce company.
# Generate synthetic data and provide insights.

np.random.seed(42)
n_customers = 500

# Generate customer data
ages = np.random.randint(18, 70, n_customers)
income = np.random.normal(50000, 20000, n_customers).clip(15000, 150000)
purchase_freq = np.random.poisson(5, n_customers)
avg_purchase_value = np.random.normal(100, 30, n_customers).clip(10, 500)

# Create DataFrame
customers = pd.DataFrame(
    {
        "age": ages,
        "income": income,
        "purchase_frequency": purchase_freq,
        "avg_purchase_value": avg_purchase_value,
    }
)

# TODO: Calculate customer lifetime value (CLV)
#       CLV = purchase_frequency * avg_purchase_value * (1 + churn_risk)
#       where churn_risk = 1 - (purchase_frequency / max_frequency)

max_frequency = customers["purchase_frequency"].max()
customers["churn_risk"] = 1 - (customers["purchase_frequency"] / max_frequency)

customers["CLV"] = (
    customers["purchase_frequency"]
    * customers["avg_purchase_value"]
    * (1 + customers["churn_risk"])
)

# print(customers[['purchase_frequency', 'avg_purchase_value', 'churn_risk', 'CLV']].head())

# TODO: Create age groups: 18-25, 26-35, 36-50, 51-70
bins = [18, 25, 35, 50, 70]
labels = ["18-25", "26-35", "36-50", "51-70"]

customers["age_group"] = pd.cut(
    customers["age"], bins=bins, labels=labels, include_lowest=True
)

# print(customers['age_group'].value_counts())

# TODO: For each age group, calculate:
#       - Number of customers
#       - Average income
#       - Average CLV
#       - Total CLV

age_group_stats = (
    customers.groupby("age_group", observed=True)
    .agg(
        num_customers=("age", "count"),
        avg_income=("income", "mean"),
        avg_CLV=("CLV", "mean"),
        total_CLV=("CLV", "sum"),
    )
    .round(2)
)

print("\n=== Stats by Age Group ===")
print(age_group_stats)

# TODO: Identify top 10% of customers by CLV
# Calculate the 90th percentile threshold
clv_threshold = customers["CLV"].quantile(0.90)
print(f"\nCLV threshold for top 10%: {clv_threshold:.2f}")

# Filter customers above that threshold
top_customers = customers[customers["CLV"] >= clv_threshold]
print(f"Number of top customers: {len(top_customers)}")
print(
    top_customers[["age", "income", "purchase_frequency", "CLV"]]
    .sort_values("CLV", ascending=False)
    .head(10)
)

# TODO: Create visualizations:
#       1. Scatter plot of income vs CLV (color by age group)
#       2. Bar chart of average CLV by age group
#       3. Correlation heatmap

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle("Customer Analysis Dashboard", fontsize=16, fontweight="bold")

age_groups_list = ["18-25", "26-35", "36-50", "51-70"]
colors = ["#e74c3c", "#3498db", "#2ecc71", "#f39c12"]

for group, color in zip(age_groups_list, colors):
    subset = customers[customers["age_group"] == group]
    axes[0].scatter(
        subset["income"], subset["CLV"], label=group, alpha=0.5, color=color, s=20
    )

axes[0].set_title("Income vs CLV by Age Group")
axes[0].set_xlabel("Income ($)")
axes[0].set_ylabel("Customer Lifetime Value ($)")
axes[0].legend(title="Age Group")

sns.barplot(
    data=age_group_stats.reset_index(),
    x="age_group",
    y="avg_CLV",
    ax=axes[1],
    palette="viridis",
)
axes[1].set_title("Average CLV by Age Group")
axes[1].set_xlabel("Age Group")
axes[1].set_ylabel("Average CLV ($)")

numerical_cols = customers[
    ["age", "income", "purchase_frequency", "avg_purchase_value", "CLV"]
]
sns.heatmap(numerical_cols.corr(), annot=True, fmt=".2f", cmap="coolwarm", ax=axes[2])
axes[2].set_title("Correlation Matrix")

plt.tight_layout()
plt.show()

# TODO: Write a brief analysis paragraph (as a markdown cell)
#       summarizing key findings and recommendations

analysis = """
The dataset consists of 500 customers analyzed across four age groups. 
The analysis of Customer Lifetime Value (CLV) shows that the average CLV across all customers is $743.76, 
while the top 10% CLV threshold is $1,222.94. The age group that generates the highest average CLV is 51-70. 
Analysis indicated that customers with higher purchase frequency tend to have lower churn risk and therefore higher CLV. 
Additionally, the 51-70 age group contributes the most value on average. 
Also, income alone does not strongly predict CLV, as customer purchase behaviour plays a more significant role. 
Thus, it is recommended that retention efforts be focused on low-frequency buyers who are at higher risk of churn, 
while the 51-70 age group should be targeted for premium product promotions. 
Furthermore, implementing loyalty programs that incentivize increased purchase frequency could directly improve overall CLV.
"""
print(analysis)
```
